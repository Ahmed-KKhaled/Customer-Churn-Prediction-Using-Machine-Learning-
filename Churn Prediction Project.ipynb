```python
important_cat = df_mi.head().index.to_list()
important_cat
```

```python
df_train['tenure'] = np.log1p(df_train['tenure'])
df_train['monthlycharges'] = np.log1p(df_train['monthlycharges'])
df_train['totalcharges'] = np.log1p(df_train['totalcharges'])

df_val['tenure'] = np.log1p(df_val['tenure'])
df_val['monthlycharges'] = np.log1p(df_val['monthlycharges'])
df_val['totalcharges'] = np.log1p(df_val['totalcharges'])

df_test['tenure'] = np.log1p(df_test['tenure'])
df_test['monthlycharges'] = np.log1p(df_test['monthlycharges'])
df_test['totalcharges'] = np.log1p(df_test['totalcharges'])
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer

def preprocessiong(df_train, df_val, df_test, cat, num):
    ohe = OneHotEncoder(drop = 'first')
    scalar = StandardScaler()
    
    transformer = make_column_transformer((scalar, num),(ohe, cat), remainder='passthrough',verbose_feature_names_out=False)
    
    X_train = transformer.fit_transform(df_train[num+cat])
    X_val = transformer.transform(df_val[num + cat])
    X_test = transformer.transform(df_test[num + cat])
    columns = transformer.get_feature_names_out()
    
    return X_train, X_val, X_test, columns
    
```

```python
X_train, X_val, X_test, columns = preprocessiong(df_train, df_val, df_test, cat_cols, num_cols)
```

```python
X_train
```

```python
y_train = df_train['churn']
y_val = df_val['churn']
y_test = df_test['churn']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)

X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

# LogisticRegression model

```python
from sklearn.linear_model import LogisticRegression
model_1 = LogisticRegression(solver='liblinear', C=0.1, max_iter=100, class_weight='balanced')
model_1.fit(X_train_res, y_train_res)  # ====> Weights
```
